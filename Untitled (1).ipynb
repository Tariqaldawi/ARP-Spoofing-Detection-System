```python
x = data[['ARP_Req_Count','ARP_Rep_Count','ARP_Ratio','ARP_Rate','ARP_Avg_IAT','Max_MAC_Freq','Unique_IPs']]
y = data['Label']
```

```python
x_train,x_test,y_train,y_test  = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y, shuffle=True)
```

# model 1

```python
model_1 = RandomForestClassifier()
```

```python
model_1.fit(x_train,y_train)
```

```python
pred_model1 = model_1.predict(x_test)
pred_model1[0]
```

# model 2

```python
model_2 = LogisticRegression()
```

```python
model_2.fit(x_train,y_train)
```

```python
pred_model2 = model_2.predict(x_test)
pred_model2[0]
```

#  

```python
print("Random Forest:\n", classification_report(y_test, pred_model1))
print("Logistic Regression:\n", classification_report(y_test, pred_model2))
```

```python
print("Random Forest:\n", accuracy_score(y_test, pred_model1))
print("Logistic Regression:\n", accuracy_score(y_test, pred_model2))
```
